# Visualizations of new Jaccard distance data

```python
#Read in packages
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.style.use('fivethirtyeight')
matplotlib.rc('xtick', labelsize=12) 
matplotlib.rc('ytick', labelsize=12)
```

```python
#Rename this to the relevant path to jaccard_distances_raw folder
path = "/Users/rpark/Documents/UChicago/Year 3/Mathematical Methods for Biological Sciences II/CP_2023/notebooks/jaccardDistort/ourdata/jaccard_out/"
```

```python
#Import ambient files, calculate means for all by reduction method
exAmb = pd.read_csv(path+'exAmb.csv',index_col=0).groupby('latent').mean()
exAmb['dataset'] = "Mouse Ex-Utero E8.5 Data"
rcgAmb = pd.read_csv(path+'rcgAmb.csv',index_col=0).groupby('latent').mean()
rcgAmb['dataset'] = "Mouse Retinal Ganglion Cell Data"
squamAmb = pd.read_csv(path+"squamAmb.csv",index_col=0).groupby('latent').mean()
squamAmb['dataset']="Human Squamous Cell Carcinoma Data"
tenxAmb = pd.read_csv(path+"tenxAmb.csv",index_col=0).groupby('latent').mean()
tenxAmb['dataset']="Mouse 10X VMH Neurons Data"
merAmb = pd.read_csv(path+"merAmb.csv",index_col=0).groupby('latent').mean()
merAmb['dataset']="MERFISH Mouse Primary Cortex Data"
smartAmb = pd.read_csv(path+"smartAmb.csv",index_col=0).groupby('latent').mean()
smartAmb['dataset']="SMART-Seq VMH data"
striatAmb = pd.read_csv(path+"striatAmb.csv",index_col=0).groupby('latent').mean()
striatAmb['dataset']="Mouse Striatum data"
```

```python
#Merge into one file for plotting
ambient = exAmb.append(rcgAmb).append(squamAmb).append(tenxAmb).append(merAmb).append(smartAmb).append(striatAmb).reset_index()
ambient.head()
```

```python
#Group data by PC number
ambient100 = ambient.loc[ambient.latent.isin(['PCA 100D','PCA 100D UMAP','PCA 100D t-SNE'])]
ambient15 = ambient.loc[ambient.latent.isin(['PCA 15D','PCA 15D UMAP','PCA 15D t-SNE'])]
ambient50 = ambient.loc[ambient.latent.isin(['PCA 50D','PCA 50D UMAP','PCA 50D t-SNE'])]
ambient2 =  ambient.loc[ambient.latent.isin(['PCA 2D','UMAP 2D','t-SNE 2D'])]
```

```python
#scatterplots, may want to make boxplot?
fig, ax = plt.subplots(2, 2,figsize = (15,10))
fig.set_facecolor('white')
ax[0,0].scatter(ambient2.x,ambient2.latent,color='red')
ax[0,0].set_xlim(0,1)
ax[0,0].set_xlabel('Jaccard Distance',fontsize = 12)
ax[0,0].set_facecolor('white')

ax[0,1].scatter(ambient15.x,ambient15.latent,color = 'green')
ax[0,1].set_xlim(0,1)
ax[0,1].set_xlabel('Jaccard Distance',fontsize = 12)
ax[0,1].set_facecolor('white')

ax[1,0].scatter(ambient50.x,ambient50.latent)
ax[1,0].set_xlim(0,1)
ax[1,0].set_xlabel('Jaccard Distance',fontsize = 12)
ax[1,0].set_facecolor('white')

ax[1,1].scatter(ambient50.x,ambient100.latent,color = 'purple')
ax[1,1].set_xlim(0,1)
ax[1,1].set_xlabel('Jaccard Distance',fontsize = 12)
ax[1,1].set_facecolor('white')

fig.tight_layout(pad=4.0)
#fig.suptitle('Mean kNN (k=30) Jaccard Distances (1-Jaccard Similarity) to \nAmbient Space Across Various Dimensionality Reductions for 7 Datasets',
#            fontsize=18)
plt.show()
```

```python
#Import PCA50 data
exPCAAmb = pd.read_csv(path+"exPCAAmb.csv",index_col=0).groupby('latent').mean()
exPCAAmb['dataset'] = "Mouse Ex-Utero E8.5 Data"
rcgPCAAmb = pd.read_csv(path+'rcgPCAAmb.csv',index_col=0).groupby('latent').mean()
rcgPCAAmb['dataset'] = "Mouse Retinal Ganglion Cell Data"
squamPCAAmb = pd.read_csv(path+"squamPCAAmb.csv",index_col=0).groupby('latent').mean()
squamPCAAmb['dataset']="Human Squamous Cell Carcinoma Data"
tenxPCAAmb = pd.read_csv(path+"tenxPCAAmb.csv",index_col=0).groupby('latent').mean()
tenxPCAAmb['dataset']="Mouse 10X VMH Neurons Data"
merPCAAmb = pd.read_csv(path+"merPCAAmb.csv",index_col=0).groupby('latent').mean()
merPCAAmb['dataset']="MERFISH Mouse Primary Cortex Data"
smartPCAAmb = pd.read_csv(path+"smartPCAAmb.csv",index_col=0).groupby('latent').mean()
smartPCAAmb['dataset']="SMART-Seq VMH data"
striatPCAAmb = pd.read_csv(path+"striatPCAAmb.csv",index_col=0).groupby('latent').mean()
striatPCAAmb['dataset']="Mouse Striatum data"
```

```python
#Merge datasets
PCAambient = exPCAAmb.append(rcgPCAAmb).append(squamPCAAmb).append(tenxPCAAmb).append(merPCAAmb).append(smartPCAAmb).append(striatPCAAmb).reset_index()
```

```python
#Group data by PC number
ambient15 = PCAambient.loc[PCAambient.latent.isin(['PCA 15D UMAP','PCA 15D t-SNE'])]
ambient50 = PCAambient.loc[PCAambient.latent.isin(['PCA 50D UMAP','PCA 50D t-SNE'])]
ambient100 = PCAambient.loc[PCAambient.latent.isin(['PCA 100D UMAP','PCA 100D t-SNE'])]
```

```python
#Plot
fig,ax = plt.subplots(nrows=1, ncols=1)
fig.set_facecolor('white')
ax.set_facecolor('white')
ax.scatter(PCAambient.x,PCAambient.latent)
ax.set_xlim(0,1)
#ax.set_ylabel('Dimensionality Reduction Method')
ax.set_xlabel('Jaccard Distance')
#ax.set_title("Mean kNN (k=30) Jaccard Distances (1-Jaccard Similarity) to \n50D PCA Space Across Various Dimensionality Reductions for 7 Datasets")

plt.show()
```
